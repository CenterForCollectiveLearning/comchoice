```python
def corr(rules, alternative="alternative", column="rank", method="kendall"):
    output = []
    for key in rules.keys():
        tmp = rules[key][[alternative, column]]
        tmp["rule"] = key
        output.append(tmp)

    df = pd.concat(output, axis=0).pivot(index=alternative, columns="rule", values=column)
    df = df.reset_index()

    return df.corr(method)
```
